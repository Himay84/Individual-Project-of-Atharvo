#  to find total number of marks of each student:-

```python
data['total']=data['MathScore']+data['ReadingScore']+data['WritingScore']
data['total']
```

# Avarage mark of each student:-

```python
data['average']=data['total']/3
data['average']
```

# Check student's marks are dependent on their parents education or not?

```python
gb=data.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("relation parent education and student score")
plt.show()
```

### from above we can say that student's marks are effected with their parents education.

#  In this way we can check marks are effected with anything or not?

```python
gb1=data.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb1
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("relation between parentmaritalstatus and student score")
plt.show()
```

### from above we can say that student's marks are not effected with their Parents Marital Status.

```python
gb3=data.groupby("WklyStudyHours").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb3
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb3,annot=True)
plt.title("relation between weeklystudyhours and student score")
plt.show()
```

### from above we can say that student's marks are  effected with their Weekly Study Hours.

```python
gb5=data.groupby("PracticeSport").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb5)
plt.figure(figsize=(4,4))
sns.heatmap(gb5,annot=True)
plt.title("relation between practice sport and student score")
plt.show()
```

# from above we can say that student's marks are  effected with their sports.

```python
gb6=data.groupby("Gender").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb6)
plt.figure(figsize=(4,4))
sns.heatmap(gb6,annot=True)
plt.title("relation between Gender and student score")
plt.show()
```

### from above we can say that student's marks are  effected with their gender also.
